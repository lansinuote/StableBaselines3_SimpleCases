```python
import gym


#å®šä¹‰ç¯å¢ƒ
class MyWrapper(gym.Wrapper):

    def __init__(self):
        env = gym.make('CartPole-v1')
        super().__init__(env)
        self.env = env

    def reset(self):
        state, _ = self.env.reset()
        return state

    def step(self, action):
        state, reward, done, _, info = self.env.step(action)
        return state, reward, done, info


env = MyWrapper()

env.reset()
```

```python
from stable_baselines3.common.env_util import make_vec_env
from sb3_contrib import TRPO

#sb3_contribæä¾›äº†ä¸€äº›ç®—æ³•,åˆ—è¡¨:https://sb3-contrib.readthedocs.io/en/master/guide/examples.html#tqc
#å„ä¸ªç®—æ³•çš„é€‚ç”¨ç¯å¢ƒ:https://stable-baselines3.readthedocs.io/en/master/guide/algos.html
model = TRPO(policy='MlpPolicy', env=env, verbose=0)

model
```

```python
from stable_baselines3.common.evaluation import evaluate_policy

#æµ‹è¯•
evaluate_policy(model, env, n_eval_episodes=10, deterministic=False)
```
